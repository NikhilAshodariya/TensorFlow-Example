```python
# importing libraries
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cross_validation import train_test_split
```

```python
from random import random
```

```python
def toList(df):
    idx =df.index.tolist()
    columns=df.columns
    l=[]
    for i in idx:
        temp=[]
        for j in columns:
            a=df.get_value(i,j)
            temp.append(a)
        l.append(temp)
    return l
```

```python
# parameters
learning_rate = 0.001
training_epochs = 250
```

```python
# loading the dataset
df = pd.read_csv('../../Data/parkinson.csv')
```

```python
df.sample()
```

```python
df.drop('subject#',axis=1,inplace=True)
```

```python
features=df.columns[:-1]
target=df.columns[-1]
```

```python
train,test=train_test_split(df,train_size=0.8)
```

```python
train_X=train[features]
train_y=train[target]
```

```python
test_X=test[features]
test_y=test[target]
```

```python
n_samples=train_X.shape
```

```python
n_samples=n_samples[0]
```

```python
n_Train_X = toList(train_X)
n_Train_y = toList(train_y.to_frame())
```

```python
n_Test_X = toList(test_X)
n_Test_y = toList(test_y.to_frame())
```

### Creating TensorFlow graph

```python
train_y.shape
```

```python
# creating input
XPlaceholder=tf.placeholder(tf.float32,[None, len(features)])
yPlaceholder=tf.placeholder(tf.float32)
```

```python
W = tf.get_variable("weigth",[len(features),1],initializer=tf.random_normal_initializer())
b = tf.get_variable("bias",(1,1),initializer=tf.random_normal_initializer())
```

```python
# constructing a linear model
pred = tf.add(tf.matmul(XPlaceholder,W),b)
```

```python
# calculating error
cost=tf.reduce_sum(tf.pow(tf.subtract(pred,yPlaceholder),2))/n_samples
```

```python
# creating a optimizer
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
sess=tf.Session()
```

```python
# launching the graph
# with tf.Session() as sess:
sess.run(tf.global_variables_initializer())

# Fit all training data
for epoch in range(training_epochs):
    for (x, y) in zip(n_Train_X, n_Train_y):
        x=np.asmatrix(x)
        sess.run(optimizer, feed_dict={XPlaceholder: x, yPlaceholder: y[0]})

    sess.run(W)
    sess.run(b)
val=sess.run(pred,feed_dict={XPlaceholder:n_Test_X})
    
    
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr=lr.fit(train_X,train_y)
```

```python
pr=lr.predict(test_X)
```

```python
# calculating error
np.sum(np.power(np.subtract(pr,val),2))/n_samples
```
